```python
import pandas as pd
import numpy as np
import matplotlib


df = pd.read_csv("C:/Users/Juuso/amk_projects/ohjelmistoprojekti2/datasetcleaned.csv")

#filter to games in Arabia map
table_arabia = df.loc[df['map_type.name']  == 'Arabia']

#wins with civ / all games with civ
arabia_winrates = table_arabia['civ.win.name'].value_counts()/(table_arabia['civ.win.name'].value_counts()+table_arabia['civ.lose.name'].value_counts())
#sort winrates and name columns
arabia_winrates = arabia_winrates.sort_values(ascending = False).reset_index().rename(columns={'index':'civs',0:'win_rate'})
#winrate to percentage
arabia_winrates.win_rate = arabia_winrates.win_rate * 100
#index starts from 1
arabia_winrates.index += 1

print('Winrates of civs in Arabia:')
print(arabia_winrates , '\n')
print('The civilization that wins most often Arabia matches is', arabia_winrates.civs[1], '\n')

print('Top 10 civs in Arabia by amount of won matches plotted on a piechart:')
data_plot = table_arabia['civ.win.name'].value_counts().iloc[0:10].plot.pie()
```
